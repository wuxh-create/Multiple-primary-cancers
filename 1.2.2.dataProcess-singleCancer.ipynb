```python
### 读取所有白人多原发癌样本
mpcDict = {}
with open('/home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/03.multiPcancer/multiPCancer.sample',"r") as fi:
    for line in fi:
        sampleID = line.strip()
        mpcDict[sampleID] = 1
fi.close()

### 提取白人单癌样本
fo = open("/home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/08.singleCancer/singleCacner.filtered.white.sample","w")
with open("/home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/04.cancerGroup/03.whiteData/all.sample","r") as fi:
    for line in fi:
        sampleID = line.strip()
        if sampleID not in mpcDict:
            fo.write(sampleID+"\n")
fi.close()
fo.close()
```
