```python
import xlrd
import xlwt
import numpy as np
```

```python
# 创建excel表格类型文件
book = xlwt.Workbook(encoding='utf-8', style_compression=0)
sheet = book.add_sheet('lung-digestive', cell_overwrite_ok=True)
with open('/home/luohh/UKB50wMultiPcancer/03.result/02.Lung-Digestive/sigGWA.5e-6.header.results',"r") as fi:
    head = fi.readline()
    header = head.strip().split()
    for n in range(len(header)):
        sheet.write(0, n, header[n])
    nrow = 0
    lines = fi.readlines()
    for i in range(len(lines)):
        nrow += 1
        data = lines[i].strip().split()
        for k in range(len(data)):
            sheet.write(nrow, k, data[k])
fi.close()

book.save('/home/luohh/UKB50wMultiPcancer/03.result/02.Lung-Digestive/pcancer.GWA5e-6.results.xls')
```
