```python
cancerList = ["anus","bone","brain","breast","cervix","colorectal","esophagus","eyeAndOrbit","gallbladderAndBiliaryTract","headAndNeck",
            "kaposiSarcoma","kidney","liver","lung","lymphoidNeoplasms","melanoma","mesothelioma","myeloidNeoplasms","otherDigestive",
           "otherEndocrine","otherFemaleGenital","otherMaleGenital","otherNervousSystem","otherRespiratory","otherUrinaryOrgans",
           "ovary","pancreas","prostate","smallIntestine","softTissueSarcoma","stomach","tCellAndNKCellNeoplasms","testis","thyroid",
           "urinaryBladder","uterus"]
```

```python
### 拷贝index MPC和所有癌症
os.system("mkdir -p /home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/01.data/01.cancerSample")
os.system("mkdir -p /home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/01.data/02.indexMPCSample")
os.system("cp /home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/04.cancerGroup/03.whiteData/*.filtered.white.sample /home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/01.data/01.cancerSample")
os.system("cp /home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/07.SpecificMPC/* /home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/01.data/02.indexMPCSample")
```

```python
### 统计每种index 癌症和多原发癌的数量及比例
fo = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/03.result/2.indexCancer.NumAndProportion.data","w")
fo.write("\t".join(["cancerName","mpcNumber","singleCancerNumber","totalNumber","proportion"])+"\n")
for cancer in cancerList:
    # 多原发癌数量
    fi = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/01.data/02.indexMPCSample/"+cancer+".mpc.sample","r")
    mpcNum = len(fi.readlines())
    fi.close()
    
    # index癌症所有癌症样本数量
    fi = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/01.data/01.cancerSample/"+cancer+".filtered.white.sample","r")
    totalNum = len(fi.readlines())
    fi.close()
    
    # 单癌样本数量
    singleCNum = totalNum-mpcNum
    
    # mpc所占比例
    proportion = mpcNum/totalNum*100
    
    fo.write("\t".join([cancer,str(mpcNum),str(singleCNum),str(totalNum),str(proportion)])+"\n")
fo.close()
```
