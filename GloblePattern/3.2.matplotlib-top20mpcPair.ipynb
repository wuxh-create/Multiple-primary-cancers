```python
import matplotlib.pyplot as plt
```

```python
top20PairDict = {}
with open('/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/03.result/3.1.number.Top20.cancerpair','r') as fi:
    for line in fi:
        data = line.strip().split("\t")
        top20PairDict[data[0]] = int(data[1])
fi.close()
```

```python
plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 14
plt.rcParams['font.weight'] = 'normal'

from matplotlib.font_manager import FontProperties
# FontProperties
labelfont = FontProperties(family='Arial',weight="bold",size=12)

colors = "#C70039"

x = list(reversed(list(top20PairDict.keys())))
y = list(reversed(list(top20PairDict.values())))

fig, ax = plt.subplots(figsize=(6,8),dpi=600)
ax.barh(x, y,align='center',color=colors)
ax.set(xlim=[-3,480],ylim=[-0.5,19.5])
ax.set_xlabel("Number of individuals",fontsize=16,fontweight="bold")

for x,y in zip(x,y):
    ax.text(y,x,s=y,size=14,va='center')

plt.savefig("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/04.plot/3.top20TimeSorted.pdf",bbox_inches='tight',format="pdf")
```
