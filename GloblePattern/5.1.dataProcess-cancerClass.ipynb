```python
cancerList = ["anus","bone","brain","breast","cervix","colorectal","esophagus","eyeAndOrbit","gallbladderAndBiliaryTract","headAndNeck",
            "kaposiSarcoma","kidney","liver","lung","lymphoidNeoplasms","melanoma","mesothelioma","myeloidNeoplasms","otherDigestive",
           "otherEndocrine","otherFemaleGenital","otherMaleGenital","otherNervousSystem","otherRespiratory","otherUrinaryOrgans",
           "ovary","pancreas","prostate","smallIntestine","softTissueSarcoma","stomach","tCellAndNKCellNeoplasms","testis","thyroid",
           "urinaryBladder","uterus"]
```

```python
countDict = {}
for cancer in cancerList:
    countDict[cancer] = {}

### 读取原始数据
with open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/03.result/3.cancerPair.allPcancer.timeSorted.results","r") as fi:
    for line in fi:
        data = line.strip().split()
        cancer1 = data[0].split("-")[0]
        cancer2 = data[0].split("-")[1]
        countDict[cancer1][cancer2] = int(data[1])
fi.close()

### 获取癌症总数
totalDict = {}
for cancer in cancerList:
    n = 0
    for key,value in countDict[cancer].items():
        n += value
    totalDict[cancer] = n

### 准备输出文件
results1Dict = {}
results2Dict = {}
for cancer in cancerList:
    if totalDict[cancer] == 0:
        continue
    myList = list(countDict[cancer].items())
    myList.sort(key=lambda x:x[1],reverse=True)
    results1Dict[cancer] = []
    results2Dict[cancer] = []
    for i in myList:
        results1Dict[cancer].append(str(i[1]/totalDict[cancer]*100))
        results2Dict[cancer].append(":".join([i[0],str(i[1]/totalDict[cancer])]))

### 输出
fo = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/03.result/5.1.cancerClassPlot.data","w")
mylist = []
for i in range(1,37):
    mylist.append("cancer"+str(i))
fo.write("\t".join(["cancerName"]+mylist)+"\n")
for cancer in cancerList:
    if cancer in results1Dict:
        fo.write("\t".join([cancer]+results1Dict[cancer])+"\n")
fo.close()
                                    
fo = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/03.result/5.2.cancerClassRaw.data","w")
for cancer in cancerList:
    if cancer in results2Dict:
        fo.write("\t".join([cancer]+results2Dict[cancer])+"\n")
fo.close()
```
