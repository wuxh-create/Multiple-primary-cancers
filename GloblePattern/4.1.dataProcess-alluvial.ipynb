```python
category = ["anus","bone","brain","breast","cervix","colorectal","esophagus","eyeAndOrbit","gallbladderAndBiliaryTract","headAndNeck",
            "kaposiSarcoma","kidney","liver","lung","lymphoidNeoplasms","melanoma","mesothelioma","myeloidNeoplasms","otherDigestive",
           "otherEndocrine","otherFemaleGenital","otherMaleGenital","otherNervousSystem","otherRespiratory","otherUrinaryOrgans",
           "ovary","pancreas","prostate","smallIntestine","softTissueSarcoma","stomach","tCellAndNKCellNeoplasms","testis","thyroid",
           "urinaryBladder","uterus"]
```

```python
### 两种癌症MPC类型转换
countDict = {}
for cancer1 in category:
    countDict[cancer1+"1"] = {}
    for cancer2 in category:
        countDict[cancer1+"1"][cancer2+"2"] = 0
with open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/01.data/03.diagnosisOrder/cancerDiagnose.allMPC.2mpc.date","r") as fi:
    for line in fi:
        data = line.strip().split("\t")
        cancer1 = data[1].split(":")[0]
        cancer2 = data[2].split(":")[0]
        countDict[cancer1+"1"][cancer2+"2"] += 1
fi.close()

fo = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/03.result/4.1.alluvial.2mpc.data","w")
fo.write("\t".join(("cancer1","cancer2","value"))+"\n")
for key,value in countDict.items():
    for k,v in value.items():
        if v == 0:
            continue
        else:
            fo.write("\t".join((key,k,str(v)))+"\n")
fo.close()
```

```python
### 两种癌症MPC类型转换 (大于10)
fo = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/03.result/4.1.alluvial.2mpc.gt25.data","w")
with open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/03.result/4.1.alluvial.2mpc.data","r") as fi:
    fo.write(fi.readline())
    for line in fi:
        data = line.strip().split("\t")
        if int(data[2]) >= 25:
            fo.write(line)
fi.close()
fo.close()
```

```python
### 三种癌症MPC转化类型
countDict = {}
for cancer1 in category:
    countDict[cancer1+"1"] = {}
    countDict[cancer1+"2"] = {}
    for cancer2 in category:
        countDict[cancer1+"1"][cancer2+"2"] = 0
        countDict[cancer1+"2"][cancer2+"3"] = 0
with open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/01.data/03.diagnosisOrder/cancerDiagnose.allMPC.3mpc.date","r") as fi:
    for line in fi:
        data = line.strip().split("\t")
        cancer1 = data[1].split(":")[0]
        cancer2 = data[2].split(":")[0]
        cancer3 = data[3].split(":")[0]
        countDict[cancer1+"1"][cancer2+"2"] += 1
        countDict[cancer2+"2"][cancer3+"3"] += 1
fi.close()

fo = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/01.globlePattern/03.result/4.2.alluvial.3mpc.data","w")
fo.write("\t".join(("cancer1","cancer2","value"))+"\n")
for key,value in countDict.items():
    for k,v in value.items():
        if v == 0:
            continue
        else:
            fo.write("\t".join((key,k,str(v)))+"\n")
fo.close()
```
