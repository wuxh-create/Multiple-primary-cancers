```python
cancerList = ["anus","bone","brain","breast","cervix","colorectal","esophagus","eyeAndOrbit","gallbladderAndBiliaryTract","headAndNeck",
            "kaposiSarcoma","kidney","liver","lung","lymphoidNeoplasms","melanoma","mesothelioma","myeloidNeoplasms","otherDigestive",
           "otherEndocrine","otherFemaleGenital","otherMaleGenital","otherNervousSystem","otherRespiratory","otherUrinaryOrgans",
           "ovary","pancreas","prostate","smallIntestine","softTissueSarcoma","stomach","tCellAndNKCellNeoplasms","testis","thyroid",
           "urinaryBladder","uterus"]
```

```python
### 处理白人癌症样本
whitePeopleDict = {}
with open("/home/luohh/UKB50wMultiPcancer/01.data/02.phenoData/basic.pheno","r") as fi:
    for line in fi:
        data = line.strip().split()
        if data[4] in ["1","1001","1002","1003"]:
            whitePeopleDict[data[0]] = 1
fi.close()

for cancer in cancerList:
    fo = open("/home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/04.cancerGroup/03.whiteData/"+cancer+".filtered.white.sample","w")
    with open("/home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/04.cancerGroup/02.filteredData/"+cancer+".filtered.sample","r") as fi:
        for line in fi:
            sample = line.strip()
            if sample in whitePeopleDict:
                fo.write(sample+"\n")
    fi.close()
    fo.close()
```

```python
### 处理非癌样本
fo = open("/home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/06.nonCancer/nonCancer.filtered.white.sample","w")
with open("/home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/06.nonCancer/nonCancer.filtered.sample","r") as fi:
    for line in fi:
        sample = line.strip()
        if sample in whitePeopleDict:
            fo.write(sample+"\n")
fi.close()
fo.close()
```
