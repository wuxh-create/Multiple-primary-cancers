```python
allSingleCancerSample = {}
with open("/home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/08.singleCancer/singleCacner.filtered.white.sample","r") as fi:
    for line in fi:
        sampleID = line.strip()
        allSingleCancerSample[sampleID] = 1
fi.close()

for cancer in cancerList:
    fo = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/02.singleCancerGWAS/01.data/04.sampleData/"+cancer+".singleCancer.sample","w")
    with open("/home/luohh/UKB50wMultiPcancer/01.data/04.sampleData/04.cancerGroup/03.whiteData/"+cancer+".filtered.white.sample","r") as fi:
        for line in fi:
            sampleID = line.strip()
            if sampleID in allSingleCancerSample:
                fo.write(sampleID+"\n")
    fi.close()
    fo.close()
```
