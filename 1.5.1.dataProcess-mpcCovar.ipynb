```python
### 性别
sexDict = {}
with open("/home/luohh/UKB50wMultiPcancer/01.data/02.phenoData/basic.pheno","r") as fi:
    for line in fi:
        data = line.strip().split("\t")
        if data[1] == "male":
            sexDict[data[0]] = "M"
        elif data[1] == "female":
            sexDict[data[0]] = "F"
fi.close()
```

```python
### 人群
popDict = {}
with open("/home/luohh/UKB50wMultiPcancer/01.data/02.phenoData/basic.pheno","r") as fi:
    for line in fi:
        data = line.strip().split("\t")
        if data[4] in ["1","1001","1002","1003"]:
            popDict[data[0]] = "White"
        elif data[4] in ["2","2001","2002","2003","2004"]:
            popDict[data[0]] = "Mixed"
        elif data[4] in ["3","3001","3002","3003","3004"]:
            popDict[data[0]] = "Asian"
        elif data[4] in ["4","4001","4002","4003"]:
            popDict[data[0]] = "Black"
        elif data[4] in ["5"]:
            popDict[data[0]] = "Chinese"
        elif data[4] in ["6"]:
            popDict[data[0]] = "Other"
        else:
            popDict[data[0]] = "Unknown"
fi.close()
```

```python
### 生成 Multi P-cancer 性别协变量文件
fo = open("/home/luohh/UKB50wMultiPcancer/01.data/02.phenoData/01.MPC/GWAS.covar","w")
with open("/home/luohh/UKB50wMultiPcancer/01.data/02.phenoData/01.MPC/GWAS.sample","r") as fi:
    for line in fi:
        sample = line.strip()
        fo.write("\t".join((sample,sample,sexDict[sample]))+"\n")
fi.close()
fo.close()
```
