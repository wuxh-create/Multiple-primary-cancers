```python
### 生成表型文件
# 获取表型数据
clinicDataDict = {}
with open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/04.gwasSurvival/01.data/02.phenoData/clinic.tsv","r") as fi:
    fi.readline()
    for line in fi:
        data = line.strip().split()
        clinicDataDict[data[0]] = data
fi.close()

### 基因型协变量 SNV
snvPcaInfoDict = {}
with open("/home/luohh/UKB50wMultiPcancer/01.data/02.phenoData/pca40.snv.eigenvec","r") as fi:
    for line in fi:
        data = line.strip().split("\t")
        snvPcaInfoDict[data[0]] = data[:12]
fi.close()

### 年龄
ageDict = {}
with open("/home/luohh/UKB50wMultiPcancer/01.data/02.phenoData/basic.pheno","r") as fi:
    fi.readline()
    for line in fi:
        data = line.strip().split("\t")
        ageDict[data[0]] = eval(data[3])
fi.close()
```

```python
#  GWAS survival 表型数据准备
fo = open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/04.gwasSurvival/01.data/02.phenoData/pheno.txt","w")
fo.write("\t".join(["sampleID","caseControl","AgeOfEventFinal","osYears","sex","age","ageAtDiagnosis","stage","pc1","pc2","pc3","pc4","pc5","pc6","pc7","pc8","pc9","pc10"])+"\n")
for key,value in clinicDataDict.items():
    if value[6] == "alive":
        caseControl = "0"
    if value[6] == "dead":
        caseControl = "1"
    
    if value[1] == "female":
        sex = "2"
    if value[1] == "male":
        sex = "1"
    
    if value[5] == "0":
        osYears = str(1/365)
    else:
        osYears = str(int(value[5])/365)

    
    fo.write("\t".join([key,caseControl,value[2],osYears]+[sex,str(ageDict[key]),value[7],value[8]]+snvPcaInfoDict[key][2:])+"\n")
fo.close()
```

```python
# 获取sampleid
with open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/04.gwasSurvival/01.data/02.phenoData/sampleID.txt","w") as fi:
    for key in clinicDataDict:
        fi.write(key+"\n")
fi.close()

with open("/home/luohh/UKB50wMultiPcancer/05.otherAnalysis/04.gwasSurvival/01.data/02.phenoData/sampleID.gcta.txt","w") as fi:
    for key in clinicDataDict:
        fi.write(key+"\t"+key+"\n")
fi.close()
```
